```python
import cv2
import numpy as np
import os

from PIL import Image
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D, BatchNormalization,Conv1D,MaxPooling1D
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import backend as K
from keras.optimizers import Adam,Adagrad
from tensorflow.keras.models import load_model
import numpy as np
import cvlib as cv
```

```python
train = 'D:/YL/Deep learning/Database/New folder (2)/Gender/Train'
test = 'D:/YL/Deep learning/Database/New folder (2)/Gender/Test'
```

```python
train_dataset = []
train_labels = []
test_dataset = []
test_labels = []

for path, _, files in os.walk(train):
    for file in files:
        image = cv2.imread(path + '\\' + file)
        image = cv2.resize(image, (50, 50))
        image = image.astype('float32')
        image /= 255.0
        train_dataset.append(image)
 
        label = path.split(os.path.sep)[-1]     
        if label=="Male":
            label=0
            print(label)
        else:
            label =1
            print(label)
            
        train_labels.append(label)
        
for path, _, files in os.walk(test):
    for file in files:
        image = cv2.imread(path + '\\' + file)
        image = cv2.resize(image, (50, 50))
        image = image.astype('float32')
        test_dataset.append(image)
 
        label = path.split(os.path.sep)[-1]     
        if label=="Male":
            label=0
            print(label)
        else:
            label =1
            print(label)
            
        test_labels.append(label)
```

```python
test_labels = np.array(test_labels)
test_dataset = np.array(test_dataset)
```

```python
train_labels = np.array(train_labels)
train_dataset = np.array(train_dataset)
```

```python
(trainX, testX, trainY, testY) = train_test_split(train_dataset, train_labels, test_size=0.3,random_state=50)
trainY = to_categorical(trainY, num_classes=2)
testY = to_categorical(testY, num_classes=2)
```

```python
(test_trainX, test_testX, test_trainY, test_testY) = train_test_split(test_dataset, test_labels, test_size=0.3,random_state=50)
test_trainY = to_categorical(test_trainY, num_classes=2)
test_testY = to_categorical(test_testY, num_classes=2)

```

```python
train_datagen = ImageDataGenerator(rotation_range=25, width_shift_range=0.1,
                         height_shift_range=0.1, shear_range=0.2, zoom_range=0.2,
                         horizontal_flip=True, fill_mode="nearest")

```

```python
epochs=100
```

```python
validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow(
   test_trainX, test_trainY, batch_size=64
)
```

```python
gender_model=Sequential()
gender_model.add(Convolution2D(32, (3, 3), padding='same', input_shape=(50,50,3), activation='relu'))
gender_model.add(MaxPooling2D(pool_size=(2,2)))
gender_model.add(BatchNormalization())

gender_model.add(Convolution2D(64, (3, 3), padding='same', activation='relu'))
gender_model.add(MaxPooling2D(pool_size=(2,2)))
gender_model.add(BatchNormalization())

gender_model.add(Convolution2D(64, (3, 3), padding='same', activation='relu'))
gender_model.add(MaxPooling2D(pool_size=(2,2)))
gender_model.add(BatchNormalization())

gender_model.add(Convolution2D(128, (3, 3), padding='same', activation='relu'))
gender_model.add(MaxPooling2D(pool_size=(2,2)))
gender_model.add(BatchNormalization())

gender_model.add(Convolution2D(256, (3, 3), padding='same', activation='relu'))
gender_model.add(BatchNormalization())

#S2 Flattening
gender_model.add(Flatten())

#S3 Fully connected 
gender_model.add(Dense(units=128, activation='relu'))
gender_model.add(Dropout(0.5))
gender_model.add(Dense(units=64, activation='relu'))
gender_model.add(Dropout(0.5))
gender_model.add(Dense(units=2, activation='softmax'))

gender_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
print(gender_model.summary())
```

```python
history = gender_model.fit_generator(train_datagen.flow(trainX, trainY, batch_size=64),
                        validation_data=validation_generator,
                        steps_per_epoch=len(trainX) // 64,
                        validation_steps=len(test_trainX)//64,
                        epochs=epochs, verbose=1)
```

```python
gender_model.save('D:/YL/Deep learning/gender_detection.model')
```

```python
ModelLoss, ModelAccuracy = gender_model.evaluate(trainX, trainY)

print('Test Loss is {}'.format(ModelLoss))
print('Test Accuracy is {}'.format(ModelAccuracy))
```
