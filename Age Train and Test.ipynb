```python
import cv2
import numpy as np
import os

from PIL import Image
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D, BatchNormalization,Conv1D,MaxPooling1D
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import backend as K
from keras.optimizers import Adam,Adagrad
from tensorflow.keras.models import load_model
import numpy as np
import cvlib as cv
from keras.callbacks import ReduceLROnPlateau
```

```python
age_train = 'D:/YL/Deep learning/Database/New folder (2)/Age/New folder'
age_test = 'D:/YL/Deep learning/Database/New folder (2)/Age/Test'
```

```python
agetrain_dataset = []
agetrain_labels = []
agetest_dataset = []
agetest_labels = []

for path, _, files in os.walk(age_train):
    for file in files:
        image = cv2.imread(path + '\\' + file)
        image = cv2.resize(image, (50, 50))
        image = image.astype('float32')
        image /= 255.0
        agetrain_dataset.append(image)
 
        label = path.split(os.path.sep)[-1]     
        if label=="1-13 years":
            label=0
            print(label)
        elif label=="14-18 years":
            label=1
            print(label)
        elif label=="19-24 years":
            label=2
            print(label)
        elif label=="25-40 years":
            label=3
            print(label)
        elif label=="41-60 years":
            label=4
            print(label)
        else:
            label=5
            print(label)
            
        agetrain_labels.append(label)
        
for path, _, files in os.walk(age_test):
    for file in files:
        image = cv2.imread(path + '\\' + file)
        image = cv2.resize(image, (50, 50))
        image = image.astype('float32')
        agetest_dataset.append(image)
 
        label = path.split(os.path.sep)[-1]    
        if label=="1-13 years":
            label=0
            print(label)
        elif label=="14-18 years":
            label=1
            print(label)
        elif label=="19-24 years":
            label=2
            print(label)
        elif label=="25-40 years":
            label=3
            print(label)
        elif label=="41-60 years":
            label=4
            print(label)
        else:
            label=5
            print(label)
            
        agetest_labels.append(label)
```

```python
print(len(agetest_labels))
print(len(agetrain_labels))
```

```python
agetest_labels = np.array(agetest_labels)
agetest_dataset = np.array(agetest_dataset)
```

```python
agetrain_labels = np.array(agetrain_labels)
agetrain_dataset = np.array(agetrain_dataset)
```

```python
(trainX, testX, trainY, testY) = train_test_split(agetrain_dataset, agetrain_labels, test_size=0.1,random_state=50)
trainY = to_categorical(trainY, num_classes=6)
testY = to_categorical(testY, num_classes=6)
```

```python
(test_trainX, test_testX, test_trainY, test_testY) = train_test_split(agetest_dataset, agetest_labels, test_size=0.1,random_state=50)
test_trainY = to_categorical(test_trainY, num_classes=6)
test_testY = to_categorical(test_testY, num_classes=6)
```

```python
agetrain_datagen = ImageDataGenerator(rotation_range=15, width_shift_range=0.1,
                         height_shift_range=0.1, shear_range=0.2, zoom_range=0.2,
                         horizontal_flip=True, fill_mode="nearest")
                                      
```

```python
validation_datagen = ImageDataGenerator(rescale=1./255 )

validation_generator = validation_datagen.flow(
   test_trainX, test_trainY
)
```

```python
INIT_LR = 1e-3
epochs=200
optim = Adam(lr=INIT_LR, decay=INIT_LR / epochs)
```

```python
age_model = Sequential()

age_model.add(Convolution2D(32, (3, 3), padding='same', input_shape=(50,50,3), activation='relu'))
age_model.add(MaxPooling2D(pool_size=(2,2)))
age_model.add(BatchNormalization())

age_model.add(Convolution2D(64, (3, 3), padding='same', activation='relu'))
age_model.add(MaxPooling2D(pool_size=(2,2)))
age_model.add(BatchNormalization())

age_model.add(Convolution2D(64, (3, 3), padding='same', activation='relu'))
age_model.add(MaxPooling2D(pool_size=(2,2)))
age_model.add(BatchNormalization())

age_model.add(Convolution2D(128, (3, 3), padding='same', activation='relu'))
age_model.add(MaxPooling2D(pool_size=(2,2)))
age_model.add(BatchNormalization())

age_model.add(Convolution2D(256, (3, 3), padding='same', activation='relu'))
age_model.add(BatchNormalization())
#S2 Flattening

age_model.add(Flatten())

#S3 Fully connected 
age_model.add(Dense(units=256, activation='relu'))
age_model.add(Dropout(0.5))
age_model.add(Dense(units=128, activation='relu'))
age_model.add(Dropout(0.5))

age_model.add(Dense(units=6, activation='softmax'))

age_model.compile(loss='categorical_crossentropy', optimizer=optim, metrics=['accuracy'])
```

```python
print(age_model.summary())
```

```python
history = age_model.fit_generator(agetrain_datagen.flow(trainX, trainY, batch_size=128),
                        validation_data=(testX,testY),
                        steps_per_epoch=len(trainX) // 128,
                        validation_steps=len(test_trainX)//128,
                        epochs=epochs,verbose=1)
```

```python
age_model.save('D:/YL/Deep learning/age_detection.model')
```

```python
ModelLoss, ModelAccuracy = age_model.evaluate(trainX, trainY)

print('Test Loss is {}'.format(ModelLoss))
print('Test Accuracy is {}'.format(ModelAccuracy))
```
